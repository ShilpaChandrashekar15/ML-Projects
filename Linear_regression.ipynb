## Linear Regression:

Linear regression is a statistical modeling technique used to analyze the relationship between a dependent variable and one or more independent variables. It assumes a linear relationship between the input variables and the output variable.

The goal of linear regression is to find the best-fit line that minimizes the difference between the predicted values and the actual values of the dependent variable. This line is determined by estimating the coefficients (slope and intercept) of the linear equation.

In simple linear regression, there is a single independent variable and one dependent variable. The equation of the line can be represented as:

y = mx + b

where:
- y is the dependent variable (also called the response variable or target variable),
- x is the independent variable (also called the predictor variable or feature),
- m is the slope of the line, which represents the change in y for a unit change in x,
- b is the y-intercept, which represents the value of y when x is zero.

The coefficients (m and b) are estimated using a method called least squares, which minimizes the sum of the squared differences between the predicted and actual values.

Linear regression provides insights into the relationship between variables, allows for prediction of the dependent variable based on independent variables, and can be used for interpreting the effects of the predictors on the outcome.



## steps carried out in Linear Regression.

1. Calculate the mean(x_mean,y_mean) of x and y
2. calculate  slope (m) = sum((xi - x_mean) * (yi - y_mean) for xi, yi in zip(x, y)) / sum( (xi - x_mean) ** 2 for xi in x )            # xi is each item in x.
3. calculate y-intercept (b) of the line : that is what is y value when x = 0
          y-intercept (b) = y_mean - m * x_mean
4. Now you have m and b, you can predict y for any given x.
          y = mx+b
          take user input i,e : x and predict y.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Diamond Carat dataset
diamonds_df = pd.read_csv('D:/projects for linkedin/data_sets/diamonds.csv')

# Extract the carat and price columns for linear regression
carat = diamonds_df['carat']
price = diamonds_df['price']

# Calculate the mean of carat and price
mean_carat = carat.mean()
mean_price = price.mean()
 
# Calculate the slope (m) and y-intercept (b) of the line
numerator = sum((xi - mean_carat) * (yi - mean_price) for xi, yi in zip(carat, price))
denominator = sum((xi - mean_carat) ** 2 for xi in carat)
m = numerator / denominator
b = mean_price - m * mean_carat

# Print the equation of the line
print("Equation of the line: price = {m} * carat + {b}")

# Predict the price for a new carat value
new_carat = float(input("enter diamon carat to predict price"))
predicted_price = m * new_carat + b
print("Predicted price for carat = {new_carat}: {predicted_price}")

# Visualization
plt.scatter(carat, price, color='blue', alpha=0.5, label='Diamonds')  # Scatter plot of diamonds
plt.plot(carat, [m * xi + b for xi in carat], color='red', label='Regression Line')  # Regression line plot
plt.scatter(new_carat, predicted_price, color='green', marker='x', label='Predicted Value')  # Scatter plot for predicted value
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Linear Regression on Diamond Carat Dataset')
plt.legend()
plt.show()
```
